```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)

titles = ['Stranger Things', 'The Office', 'Breaking Bad', 'Money Heist', 'Narcos', 'Black Mirror', 'Dark', 'The Crown', 'BoJack Horseman', 'Ozark']
genres = ['Sci-Fi', 'Comedy', 'Crime', 'Thriller', 'Drama', 'Sci-Fi', 'Mystery', 'Biography', 'Comedy', 'Crime']
dates = pd.date_range('2024-04-01', periods=30, freq='D')
durations = pd.to_timedelta(np.random.randint(20, 60, size=30), unit='m')  # Random durations between 20 and 60 minutes

data = {
    'Title': np.random.choice(titles, size=30),
    'Date': np.random.choice(dates, size=30),
    'Duration': durations,
    'Genre': np.random.choice(genres, size=30)
}

df = pd.DataFrame(data)
df.head()
```

```python
df['DayOfWeek'] = df['Date'].dt.day_name()
df
```

```python
binge = df.groupby(['Date', 'Title']).size().reset_index(name='EpisodesWatched')
binge[binge['EpisodesWatched'] >= 2]
```

```python
import matplotlib.pyplot as plt 
np.random.seed(0)

titles = ['Stranger Things', 'The Office', 'Breaking Bad', 'Money Heist', 'Narcos', 'Black Mirror', 'Dark', 'The Crown', 'BoJack Horseman', 'Ozark']
genres = ['Sci-Fi', 'Comedy', 'Crime', 'Thriller', 'Drama', 'Sci-Fi', 'Mystery', 'Biography', 'Comedy', 'Crime']
dates = pd.date_range('2024-04-01', periods=30, freq='D')
durations = pd.to_timedelta(np.random.randint(20, 60, size=30), unit='m')  # Random durations between 20 and 60 minutes

data = {
    'Title': np.random.choice(titles, size=30),
    'Date': np.random.choice(dates, size=30),
    'Duration': durations,
    'Genre': np.random.choice(genres, size=30)
}

df = pd.DataFrame(data)
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
df['Title'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Most Watched Shows')
plt.xlabel('Title')
plt.ylabel('Watch Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df['Genre'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('Genre Distribution')
plt.ylabel('')
plt.show()
```

```python
# Convert 'Duration' to minutes
df['Minutes'] = df['Duration'].dt.total_seconds() / 60

# Group by Title and sum minutes
watch_time = df.groupby('Title')['Minutes'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
watch_time.plot(kind='bar', color='orange')
plt.title('Total Watch Time by Show (in Minutes)')
plt.xlabel('Title')
plt.ylabel('Total Minutes Watched')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
